# 3:continent with more beer,spirit,wine servings 

```python
drinks3=drinks.copy()
```

```python
drinks3.columns
```

# -continent with more beer_servings

```python
continent=drinks3.groupby('continent').agg('sum').sort_values('beer_servings',ascending=False)

continent[:1].beer_servings
```

# -continent with more spirit_servings

```python
continent=drinks3.groupby('continent').agg('sum').sort_values('spirit_servings',ascending=False)

continent[:1].spirit_servings
```

# -continent with more wine servings

```python
continent=drinks3.groupby('continent').agg('sum').sort_values('wine_servings',ascending=False)

continent[:1].wine_servings
```

### -overall memory usage

```python
ufo.info(memory_usage='deep')
```

### OVERALL MEMORY USAGE IS 5.1 MG

## -Column with highest share of memory

```python
ufo.memory_usage(deep=True)
```

### COLUMN WITH HIGHEST SHARE OF MEMORY IS "TIME"

```python
ufo.dtypes
```

```python
ufo.memory_usage(deep=True)
```

```python
ufo.info(memory_usage='deep')
```

## i reduce the overall memory_usage by 1.1MB, as I change the time type from object to dateTime type, object occupies more memory space than dateTime
